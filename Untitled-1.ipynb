```python
import os
import pandas as pd
import numpy as np
import random
import difflib
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier 
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import warnings

warnings.filterwarnings('ignore')
sns.set_style("whitegrid")
```

```python
# Common digraphs in English
TOP_DIGRAPHS = ['th', 'he', 'in', 'er', 'an', 're', 'on', 'at', 'en', 'nd', 'ti', 'es', 'or', 'te', 'of']
# Common keys
TOP_KEYS = ['e', 't', 'a', 'o', 'i', 'n', 's', 'r', 'h', 'space'] 
```

```python
DATASET_PATH = r"D:\ML_project\Keystrokes_dataset\files"

def get_text_similarity(s1, s2):
    return difflib.SequenceMatcher(None, str(s1), str(s2)).ratio()

def extract_advanced_features(group_df):
    feats = {}
    feats['PARTICIPANT_ID'] = group_df['PARTICIPANT_ID'].iloc[0]
    
    # --- 1. Global Stats ---
    for metric in ['HOLD_TIME', 'FLIGHT_TIME', 'LATENCY']:
        feats[f'Mean_{metric}'] = group_df[metric].mean()
        feats[f'Std_{metric}'] = group_df[metric].std()
    
    # --- 2. Digraph Latencies ---
    # We need to reconstruct the sequence of keys to find digraphs
    group_df = group_df.sort_values('PRESS_TIME')
    
    # Create a column for Previous Letter to form digraphs
    group_df['PREV_LETTER'] = group_df['LETTER'].shift(1)
    group_df['DIGRAPH'] = group_df['PREV_LETTER'].astype(str) + group_df['LETTER'].astype(str)
    
    for dig in TOP_DIGRAPHS:
        first_char = dig[0]
        second_char = dig[1]
        
        next_letter = group_df['LETTER'].shift(-1)
        mask = (group_df['LETTER'] == first_char) & (next_letter == second_char)
        
        matches = group_df.loc[mask, 'LATENCY']
        if len(matches) > 0:
            feats[f'Digraph_{dig}'] = matches.mean()
        else:
            feats[f'Digraph_{dig}'] = np.nan
            
    # --- 3. Key Specific Hold Times ---
    for key in TOP_KEYS:
        key_label = key
        if key == 'space': key_label = ' '
        
        matches = group_df.loc[group_df['LETTER'] == key_label, 'HOLD_TIME']
        if len(matches) > 0:
            feats[f'Hold_{key}'] = matches.mean()
        else:
            feats[f'Hold_{key}'] = np.nan
            
    # --- 4. Similarity & Speed ---
    target_sentence = str(group_df['SENTENCE'].iloc[0]) if 'SENTENCE' in group_df.columns else ""
    user_input = str(group_df['USER_INPUT'].iloc[0]) if 'USER_INPUT' in group_df.columns else ""
    feats['Text_Similarity'] = get_text_similarity(target_sentence, user_input)
    
    total_time_ms = group_df['RELEASE_TIME'].max() - group_df['PRESS_TIME'].min()
    if total_time_ms is not pd.NA and total_time_ms > 0:
        feats['WPM'] = (len(group_df) / 5) / (total_time_ms / 60000)
    else:
        feats['WPM'] = 0
        
    return feats

def load_data_final(path):
    print(f"Scanning directory: {path}...")
    try: all_files = os.listdir(path)
    except Exception: return []

    # Group files
    file_map = {}
    for f in all_files:
        if '_' in f:
            pid = f.split('_')[0]
            file_map.setdefault(pid, []).append(f)
            
    # --- LIMIT CLASSES FOR HIGH ACCURACY (Top 20) ---
    top_participants = sorted(file_map.keys(), key=lambda x: len(file_map[x]), reverse=True)[:30]
    
    selected_files = []
    for pid in top_participants:
        selected_files.extend(file_map[pid][:200]) 
        
    print(f"Selected {len(selected_files)} files from {len(top_participants)} participants.")
    
    session_features = []
    
    for i, filename in enumerate(selected_files):
        if i % 200 == 0 and i > 0: print(f"Processed {i} files...")
        try:
            df = pd.read_csv(os.path.join(path, filename), sep='\t')
            if 'PRESS_TIME' not in df.columns: continue
            df = df.dropna(subset=['PRESS_TIME', 'RELEASE_TIME'])
            if df.empty: continue

            # Basic calc
            df['HOLD_TIME'] = df['RELEASE_TIME'] - df['PRESS_TIME']
            df['NEXT_PRESS_TIME'] = df['PRESS_TIME'].shift(-1)
            df['FLIGHT_TIME'] = df['NEXT_PRESS_TIME'] - df['RELEASE_TIME']
            df['LATENCY'] = df['NEXT_PRESS_TIME'] - df['PRESS_TIME']
            
            # Filter outliers
            df.loc[df['FLIGHT_TIME'] > 5000, 'FLIGHT_TIME'] = np.nan
            
            # Group by Sentence
            group_col = 'TEST_SECTION_ID' if 'TEST_SECTION_ID' in df.columns else 'SENTENCE'
            grouped = df.groupby(group_col)
            
            for _, group_df in grouped:
                if len(group_df) < 10: continue
                feat = extract_advanced_features(group_df)
                session_features.append(feat)
        except Exception: continue
            
    return pd.DataFrame(session_features)

df_final = load_data_final(DATASET_PATH)
print(f"Loaded {len(df_final)} samples.")
df_final.head()
```

```python
df_model = df_final.copy()

# Filter small classes
df_model = df_model.groupby('PARTICIPANT_ID').filter(lambda x: len(x) >= 10)
print(f"Final Dataset Shape: {df_model.shape}")
print(f"Classes: {df_model['PARTICIPANT_ID'].nunique()}")

X = df_model.drop(columns=['PARTICIPANT_ID'])
y = df_model['PARTICIPANT_ID']

# Impute missing values
imputer = SimpleImputer(strategy='median')
X_imputed = imputer.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42, stratify=y)

# Scale for KNN
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
print("Model Comparison")

# 1. Random Forest
rf = RandomForestClassifier(n_estimators=300, max_depth=20, min_samples_split=4, random_state=42)
rf.fit(X_train, y_train)
rf_acc = accuracy_score(y_test, rf.predict(X_test))
print(f"Random Forest Accuracy: {rf_acc:.4f}")

# 2. Decision Tree 
dt = DecisionTreeClassifier(random_state=42, max_depth=15)
dt.fit(X_train, y_train)
dt_acc = accuracy_score(y_test, dt.predict(X_test))
print(f"Decision Tree Accuracy: {dt_acc:.4f}")

# 3. KNN 
knn = KNeighborsClassifier(n_neighbors=5, weights='distance', metric='manhattan')
knn.fit(X_train_scaled, y_train)
knn_acc = accuracy_score(y_test, knn.predict(X_test_scaled))
print(f"KNN Accuracy: {knn_acc:.4f}")
```
